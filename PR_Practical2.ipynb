```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Function to compute the mean squared error
def compute_mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Function for Gradient Descent
def gradient_descent(X, y, learning_rate=0.1, n_iterations=1000):
    m = len(X)
    X_b = np.c_[np.ones((m, 1)), X]  # Add bias term (x0 = 1)
    theta = np.random.randn(2, 1)  # Random initialization of parameters
    cost_values = []

    for iteration in range(n_iterations):
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - learning_rate * gradients
        cost = compute_mse(y, X_b.dot(theta))
        cost_values.append(cost)
        print(f"Iteration: {iteration}, Theta: {theta.flatten()}, Cost: {cost:.4f}")

    return theta

# Hyperparameters
learning_rate = 0.001
n_iterations = 1000

# Run gradient descent
theta_optimal = gradient_descent(X, y, learning_rate, n_iterations)
print("Optimal parameters (theta):", theta_optimal)

# Create bar graph
width = 0.02  # Width of the bars
plt.bar(X.flatten(), y.flatten(), width=width, color='b', alpha=0.6, label='Data Points')
plt.bar(X.flatten(), (theta_optimal[0] + theta_optimal[1] * X).flatten(), width=width, color='r', alpha=0.3, label='Fitted Line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression with Gradient Descent (Bar Graph)")
plt.legend()
plt.show()
```
